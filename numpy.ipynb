### **Boolean Masking & Filtering**


```python
a = np.array([10,20,30,40,50])
print(a>25)
```

```python
a = np.array([10,20,30,40,50])

# Greater than
print("Greater than 25:")
print(a[a > 25])

# Less than
print("\nLess than 25:")
print(a[a < 25])

# Equal to
print("\nEqual to 30:")
print(a[a == 30])

# Not equal to
print("\nNot equal to 30:")
print(a[a != 30])

# Greater than or equal to
print("\nGreater than or equal to 30:")
print(a[a >= 30])

# Less than or equal to
print("\nLess than or equal to 30:")
print(a[a <= 30])

# AND condition
print("\nGreater than 20 AND less than 50:")
print(a[(a > 20) & (a < 50)])

# OR condition
print("\nLess than 20 OR greater than 40:")
print(a[(a < 20) | (a > 40)])
```

```python
marks = np.array([78,85,92,66,88])

print("Passed with distinction:")
print(marks[marks >= 80])

print("Need improvement:")
print(marks[marks < 80])

print("Top performers (80-90):")
print(marks[(marks >= 80) & (marks <= 90)])
```

```python
marks = np.array([78, 85, 92, 66, 88])

top_students = marks[marks > 80]

print("Top Students Marks:", top_students)
```

## **Linear Algebra**

```python
import numpy as np
```

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])
```

```python
#Matrix Addition
print(" Addition = \n",A+B)

#Matrix Subtraction
print("Subtraction = \n",A-B)

#Matrix Multiplication
print("Multiplication = \n",A*B)

#Transpose
print("A transpose = \n",A.T)
print("B transpose = \n",B.T)

# Determinant
print("Determinant = ",np.linalg.det(A))

#Inverse
print("Inverse = \n",np.linalg.inv(A))

#Eigenvalues and Eigenvectors
values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:",vectors)

#Rank of Matrix
print("Rank = ",np.linalg.matrix_rank(A))
```

```python
# Linear Equation

A = np.array([[2,1],
             [1,3]])

B = np.array([5,6])

x = np.linalg.solve(A,B)
print(x)
```

**LINEAR ALGEBRA**

Most Important numpy.linalg Functions
Function	Purpose

np.linalg.det() Determinant

np.linalg.inv()	Inverse

np.linalg.eig()	Eigenvalues & Eigenvectors

np.linalg.solve()	Solve linear equations

np.linalg.matrix_rank()	Matrix rank

np.linalg.norm()	Vector/Matrix norm

np.linalg.svd()	Singular Value Decomposition



```python
a = np.array([[4,2],
              [1,3]])

print("Matrix",a)

print("Determine",np.linalg.det(a))

print("Inverse",np.linalg.inv(a))

print("Transpose",a.T)
```
